### One dimensional Data Sets

```python
import numpy as np
```

```python
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
y
```

```python
x = np.arange(len(y))
```

```python
x
```

```python
plt.plot(x,y);
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

```python
y.cumsum()
```

```python
x.cumsum()
```

```python
x
```

```python
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
```

```python
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum())-1,
        np.max(y.cumsum())+1);
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), 'b',lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum())-1,
        np.max(y.cumsum())+1);
```
