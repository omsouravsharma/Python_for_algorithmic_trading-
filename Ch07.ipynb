### One dimensional Data Sets

```python
import numpy as np
```

```python
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
y
```

```python
x = np.arange(len(y))
```

```python
x
```

```python
plt.plot(x,y);
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

```python
y.cumsum()
```

```python
x.cumsum()
```

```python
x
```

```python
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
```

```python
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum())-1,
        np.max(y.cumsum())+1);
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), 'b',lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum())-1,
        np.max(y.cumsum())+1);
```

### Two Dimesional Data Sets

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
y
```

```python
plt.figure(figsize=(10,6))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw = 1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot');
```

```python
y
```

```python
y[:,0] = y[:,0]*100
y[:,0]
```

```python
plt.figure(figsize=(10,6))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw = 1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot');
```

```python
fig, ax1 =plt.subplots()
plt.plot(y[:,0], 'b', lw= 1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('Value 1st')
plt.title('A Simple Plot')

ax2 = ax1.twinx()
plt.plot(y[:,1],'g',lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('Value 2nd');
```

```python
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(y[:,0], 'b', lw= 1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('Value 1st')
plt.title('A Simple Plot')

plt.subplot(212)
plt.plot(y[:,1],'g',lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('Value 2nd');
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(y[:,0], 'b', lw= 1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('Value 1st')
plt.title('A Simple Plot')

plt.subplot(122)
plt.bar(np.arange(len(y)), y[:,1], width = 0.5, color='g', label='2nd')
plt.legend(loc=0)
plt.ylabel('Value 2nd');
```
